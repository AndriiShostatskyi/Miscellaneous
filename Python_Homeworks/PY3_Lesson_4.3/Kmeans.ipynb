```python
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.vq import whiten, kmeans2

%matplotlib inline
```

```python
def call_api(uri, result_key):
    try:
        json = requests.get(uri, headers = { 'X-Auth-Token': 'c52b7825851b4357af3e06c1888152f0' }).json()
        return json[result_key]
    except:
        print(json)
        return None
```

```python
teams = []
seasons = requests.get('http://api.football-data.org/v1/soccerseasons', headers = { 'X-Auth-Token': 'c52b7825851b4357af3e06c1888152f0' }).json()

for season in seasons:
    teams += call_api(season['_links']['teams']['href'], 'teams')
```

```python
teams_for_pd = []

for team in teams:
    
    fixtures = call_api(team['_links']['fixtures']['href'], 'fixtures')
    nlost = 0
    
    if fixtures is None:
        continue
        
    for f in fixtures:
        
        if f is None:
            continue
            
        homeTeamGoals = 0
        awayTeamGoals = 0
        
        try:
            homeTeamGoals += f['result']['goalsHomeTeam']
            awayTeamGoals += f['result']['goalsAwayTeam']
        except:
            pass
        try:
            homeTeamGoals += f['result']['halfTime']['goalsHomeTeam']
            awayTeamGoals += f['result']['halfTime']['goalsAwayTeam']
        except:
            pass
            
        if team['name'] == f['homeTeamName']:
            if homeTeamGoals < awayTeamGoals:
                nlost += 1
        else:
            if homeTeamGoals > awayTeamGoals:
                nlost += 1
            
    val = team['squadMarketValue'] 
    
    if val is None:
        pass
    else:
        val = val.replace(' €', '').replace(',', '') 
        teams_for_pd.append({'team': team['name'], 'value': int(val), 'lost': nlost})
```

```python
teams_pd = pd.DataFrame(teams_for_pd)
teams_pd
```

```python
del teams_pd['team']
```

```python
whitened = whiten(teams_pd.as_matrix())
```

```python
centroids, cluster_map = kmeans2(whitened, 5)
```

```python
colors_map = {0: 'r', 1: 'g', 2: 'b', 3: 'y', 4: 'm'}
colors = [colors_map[c] for c in cluster_map]
```

```python
teams_pd.plot.scatter(x='value', y='lost', c=colors, s=100)
```

```python
centroids, cluster_map = kmeans2(whitened, 3)
colors = [colors_map[x] for x in cluster_map]
teams_pd.plot.scatter(x='value', y='lost', c=colors, s=100)
```
